```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(10,10))
```

```python
ax.set_xlim([-5,5])
ax.set_ylim(-5,5)
ax.patch.set_edgecolor('white')  
ax.patch.set_linewidth('2')  
ax.grid(color = 'white', ls = '-.', lw = 0.2)
ax.set_xticks(np.arange(-4, 5, 1))
ax.set_yticks(np.arange(-4, 5, 1))
ax.axhline(linewidth=4, color="white")
ax.axvline(linewidth=4, color="white") 
ax.set_facecolor((0,0,0))
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')


fig
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlim([-1,3])
ax.set_ylim(-1,3)
ax.patch.set_edgecolor('white')  
ax.patch.set_linewidth('2')  
ax.grid(color = 'white', ls = '-.', lw = 0.2)
ax.set_xticks(np.arange(-1, 4, 1))
ax.set_yticks(np.arange(-1, 3, 1))
ax.axhline(linewidth=4, color="white")
ax.axvline(linewidth=4, color="white") 
ax.set_facecolor((0,0,0))
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.arrow(0, 0, 2, 1, length_includes_head = True, width = .03, head_width = 0.2, color = 'white')
ax.plot([2, 2], [1, 0], color = 'white', ls = ':')
ax.text(2, 1.2, r'$Q$', size = 15, color = 'white')
ax.text(0.2, 0.3, r'$P$', size = 15, color = 'white')
ax.text(0.8, 0.1, r'$\theta$', size = 15, color = 'salmon')
```
